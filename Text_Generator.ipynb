#Text Generator

The following tool generates the texts according to put key sentence/words. It is computer power requiring tool, so it is recommended to run in google colab so to be able to process data properely.

```python
# ! pip install transformers
```

```python
from transformers import pipeline
```

```python
gpt_neo_generator = pipeline('text-generation', model='EleutherAI/gpt-neo-1.3B')
```

```python
# results are sampled from the top 50 candidates
sentences = gpt_neo_generator("home office concept explained", do_sample=True, top_k=50, temperature=0.6, max_length=128, num_return_sequences=3)
for sentence in sentences:
  print(sentence["generated_text"])
  print("="*50)
```
